# Bioinformatics

### Drug target
Our drug target is the Sodium-dependent Dopamine transporter (DAT), which are located in regions of the brain and the peripheral nervous system. This target is family of solute carrier proteins (SLC), especially of the sodium- and chloride-dependent neurotransmitter symporters (SLC6A). The DAT plays a major role in the pathology of ADHD, narcolepsy, depression and Parkinson's disease [1-4]. In patients suffering from these disorders, the DAT activity is increased and therefore dopamine is transported intracellularly. This leads to depleted levels of dopamine in the extracellular matrix [5]. In order to treat these diseases, dopamine reuptake inhibitors (DRIs) are needed to increase the dopamine levels by inhibiting the DAT [6]. The DAT is very similar to the norepinephrine transporter (NET). Therefore, selective DRIs (sDRIs) are necessary for the treatment of the mentioned diseases by only inhibiting the dopamine reuptake. Currently, the selective DRIs that are available on the market are amineptine (N06AA19), which is an antidepressant and modafinil (N06BA07), which is a psychostimulant [7]. However, amineptine was withdrawn from the market due to reactive metabolites and modafinil was only mildly selective to the DAT. Therefore, designing a new selective ligand for the DAT is of high importance.

### Related proteins based on sequence
The dopamine transporter has a sequence length of 620 amino acids. The protein is reviewed by Uniprot and there is evidence at protein level of the existence of the DAT. The transporter has a mass of 68.5 kDa. The accession number is Q01959.

The most identical protein in other species is the SLC6A3 transporter of the Pan paniscus (Bonobo). This protein has a length of 620 amino acids. The accession number is A0A2R8ZUK6. The protein is unreviewed and inferred from homology. The mass of the transporter is 68.6 kDa.

The most identical protein in human is the sodium-dependent norepinephrine transporter (SLC6A2). This protein has a sequence length of 617 amino acids. The accession number is P23975. The transporter is reviewed by Uniprot and there is evidence at protein level of the SLC6A2. 

The SLC6A3 transporter in Bonobo species is more similar to the original target (99.8%). The SLC6A2 in human has a similarity of 67.1% to the DAT.

### Cheminformatics

After searching for the structure-activity relationship (SAR) in literature of dopamine with the DAT, we found that the pKi is typically between 200 to 500 nM. This means that at this concentration, dopamine binds effectively at the DAT [10].

SMILES ligand: C1=CC(=C(C=C1CCN)O)O

Representative molecule: PubChem CID 77001 (1 OH group on aromatic ring and a hydrochloride).

ChEMBL ID ligand: CHEMBL59
When fetching 95% similar compounds on ChEMBL, 3 compounds were retrieved. When fetching 90% similar compounds, still 3 compounds were retrieved. A lot less compounds were retrieved from the ChEMBL similarity search compared to the similarity search on PubChem. This can be explained by the fact that ChEMBL only contains drug-like molecules and PubChem also contains non-drug-like molecules and other chemicals.

ZINC ID ligand: ZINC33882

All three databases were able to provide similar molecules to dopamine. Which database is able to provide the most similar molecules highly depends on the size and content of the databases. Therefore, the impact of the threshold is not very high.

### Discussion
The results of the docking affinities suggest that the compounds do not bind well to the DAT. However, the results from machine learning showed that our proposed ligands have a better affinity for the DAT than dopamine itself. Therefore, the pKi values obtained from the docking procedure are not highly reliable. Docking did show us that most of the expected key interactions were achieved between the proposed ligands and the target which does suggest that the ligands will have a high affinity for the DAT.

The key interactions present between the designed compounds and DAT have been determined with docking and the PLIP viewer. All 5 designed compounds have a hydrophobic interaction with PHE76, VAL152, TYR156, PHE320, PHE326, VAL328 and ALA423. Interestingly, only PHE76, VAL152 and PHE326 have been described in literature to have hydrophobic key interactions with such compounds [8]. Therefore, it is uncertain whether these other key interactions actually are present. Furthermore, hydrogen bond key interactions present in almost each of the designed compounds are from ASP79, SER149, and SER422. However, only ASP476 has been described as a hydrogen bond donor in the literature [9]. Therefore, the hydrogen bond donor key interactions between the designed molecules and DAT are still quite uncertain. However, it is well possible that SER149 and SER422 have such key interaction.
Lastly, almost each designed compound has a π-stacking interaction with TYR156 and PHE326. Even though, TYR156 is the only amino acid described in literature to have such interaction, PHE326 seems to be very important as well according to docking [9].

Interestingly, a salt bridge with ASP79 has not been shown to be present in any of the designed compounds with the docking procedure. It is likely that this happened due to the fact that the amine is not protonated in the molecules, because the protonated amine is fundamental in forming a salt bridge.

In comparison to the previously obtained biological data (pIC50) of these designed compounds, the estimated pKi values with machine learning and docking do not match. The 5 designed compounds have been biologically evaluated to determine their potency to the DAT. From compound 1 to 5, their potency are 8.3 ± 0.1, 7.8 ± 0.0, 8.2 ± 0.1, 7.5 ± 0.0​, and 6.6 ± 0.0, respectively. Although pIC50 and pKi is not directly the same, they are correlated. The estimated pKi values of the machine learning were not very diverse, which is not the case with the biological data of the designed compounds. Therefore, the pKi values of machine learning and docking are not very reliable.

# Conclusion & Future Prospects
The aim of this project was to design a new proposed ligand for the dopamine transporter. We tested 5 aromatic cycloalkylamines designed during a previous project via machine learning and docking. The results of the machine learning model showed that all compounds have a better affinity for the DAT than the reference ligand dopamine. However, the pKi values were all very close to each other and therefore the machine learning model was not able to predict which compound has the best affinity for the DAT. Furthermore, all of our proposed compounds were docked in the right binding pocket. This gave more insight in important key interactions between the ligands and target than was already described in literature. This information can be used for the design of future sDRIs for the DAT.

For future research on docking these compounds, it is necessary to protonate the amine group in every designed compound, because this is fundamental for the salt bridge key interaction with ASP79. Furthermore, AlphaFold should be used for future docking studies to obtain more information about the key interactions between ligand and target and identify the most essential functional groups, because of its greater quality. Moreover, a future study could be to dock our ligands in the NET to show whether our compounds are selective to the DAT. Therefore, more information could be obtained about which functional groups are necessary in the compounds to only be selective for the DAT.
